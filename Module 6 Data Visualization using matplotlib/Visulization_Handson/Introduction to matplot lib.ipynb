# how to create an line plot

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline


#definfing an dataset
x = np.arange(0,20,0.1)
y = 2*x + 5


#ploting an datapoints
plt.plot(x,y)
```

# Customizing an line plot


```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline


#definfing an dataset
x = np.arange(0,20,0.1)
y = 2*x + 5


plt.title("Line plot Demo")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
#ploting an datapoints
plt.plot(x,y,linewidth=2.0,linestyle=":",color="r",alpha=0.5,marker ="o")
```
