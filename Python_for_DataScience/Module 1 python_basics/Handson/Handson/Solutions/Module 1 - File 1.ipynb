```python
#q6
#1st part- checking which values are greater
times=int(input('How many numbers you want to compare? '))
e=[]
for i in range(times):
    numbers=int(input('Enter the numbers one by one: '))
    e.append(numbers)
for j in e:
    if(j==max(e)):
        print('The greatest value among the numbers is',j)
```

```python
#2nd part- amstrong number:
num = int(input("Enter a number: "))
sum = 0
temp = num
while temp > 0:
    digit = temp % 10
    sum += digit ** 3
    temp //= 10

if num == sum:
    print(num,"is an Armstrong number")
else:
    print(num,"is not an Armstrong number")
```

```python
#3rd part- 
num=int(input("Enter a number: "))
prime=True
if(num>1):
    for i in range(2,num):
        if(num%i==0):
            prime=False
    if(prime):
        print(num,' is prime.')
```

```python
#q7: 
def addition(a,b):
    return a+b
def subtraction(a,b):
    return a-b
def multiplication(a,b):
    return a*b
def division(a,b):
    return a/b
n1=int(input('Enter the 1st number'))
n2=int(input('Enter the 2nd number'))
op=input('Enter which task you need to perform')
if(op=='add' or op=='Add' or op=='Addition' or op=='addition'):
    op1=addition(n1,n2)
    print(op1) 
elif(op=='sub' or op=='Sub' or op=='Subtraction' or op=='subtraction'):
    op2=subtraction(n1,n2)
    print(op2)
elif(op=='mul' or op=='Mul' or op=='Multiplication' or op=='multiplication'):
    op3=multiplication(n1,n2)
    print(op3)
elif(op=='div' or op=='Div' or op=='Division' or op=='division'):
    op4=division(n1,n2)
    print(op4)
```

```python
#q8
nterms = int(input("How many terms? "))
n1, n2 = 0, 1
count = 0
# check if the number of terms is valid
if nterms <= 0:
    print("Please enter a positive integer")
# if there is only one term, return n1
elif nterms == 1:
    print("Fibonacci sequence upto",nterms,":")
    print(n1)
# generate fibonacci sequence
else:
    print("Fibonacci sequence:")
    while count < nterms:
        print(n1)
        nth = n1 + n2
       # update values
        n1 = n2
        n2 = nth
        count += 1
```
